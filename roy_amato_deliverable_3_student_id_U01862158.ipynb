```python
# import
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# get data
train_path = '/content/sample_data/california_housing_train.csv'
test_path = '/content/sample_data/california_housing_test.csv'
train = pd.read_csv(train_path)
test = pd.read_csv(test_path)

# define X and y
X_train = train.iloc[:, train.columns!='median_house_value']
y_train = train['median_house_value']
X_test = test.iloc[:, test.columns!='median_house_value']
y_test = test['median_house_value']

# import
import tensorflow as tf

# define a neural network model
model = tf.keras.models.Sequential(name='this_model')
model.add(tf.keras.layers.Dense(1, input_shape=[8]))
model.summary()

# compile
model.compile(optimizer='rmsprop', loss='mae')

# fit
model.fit(X_train, y_train, validation_split=0.2, epochs=10)

# save
model.save('tmp_model_yin.h5')

# predict on test set
y_test_pred_ = model.predict(X_test)

# how accurate?
# step 1: take the difference
y_test_pred_.reshape((-1)) - np.asarray(y_test)

# step 2: take the absolute value of the difference
np.abs(y_test_pred_.reshape((-1)) - np.asarray(y_test))

# sttep 3: take the average of the absolute difference
np.mean(np.abs(y_test_pred_.reshape((-1)) - np.asarray(y_test)))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Generate synthetic data ---
np.random.seed(0)
X = np.linspace(0, 10, 100)
Y = linear_regression(X, beta_0=5, beta_1=2)

# --- Gradient Descent Implementation ---
def gradient_descent(X, y, alpha=0.001, iterations=10000):
    n = len(y)
    beta_0 = 0
    beta_1 = 0
    for _ in range(iterations):
        y_pred = beta_0 + beta_1 * X
        d_beta_0 = -(2/n) * np.sum(y - y_pred)
        d_beta_1 = -(2/n) * np.sum((y - y_pred) * X)
        beta_0 -= alpha * d_beta_0
        beta_1 -= alpha * d_beta_1
    return beta_0, beta_1

# --- Analytical Solution ---
def calculate_coefficients(X, Y):
    mean_x = np.mean(X)
    mean_y = np.mean(Y)
    beta_1 = np.sum((X - mean_x) * (Y - mean_y)) / np.sum((X - mean_x) ** 2)
    beta_0 = mean_y - beta_1 * mean_x
    return beta_0, beta_1

# --- Compute coefficients ---
b0_gd, b1_gd = gradient_descent(X, Y)
b0_analytical, b1_analytical = calculate_coefficients(X, Y)

print("Gradient Descent: beta_0 =", b0_gd, ", beta_1 =", b1_gd)
print("Analytical: beta_0 =", b0_analytical, ", beta_1 =", b1_analytical)

# --- Plot ---
plt.scatter(X, Y, alpha=0.5, label="Data")
plt.plot(X, b0_gd + b1_gd*X, color="green", label="Gradient Descent")
plt.plot(X, b0_analytical + b1_analytical*X, color="red", linestyle="--", label="Analytical")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gradient Descent vs Analytical Solution")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# --- Generate synthetic data ---
np.random.seed(0)
X = np.linspace(0, 10, 50)
Y = linear_regression(X, beta_0=5, beta_1=2)

# --- Define Mean Squared Error ---
def mse(beta_0, beta_1, X, Y):
    Y_pred = beta_0 + beta_1 * X
    return np.mean((Y - Y_pred) ** 2)

# --- Create a grid of beta_0 and beta_1 values ---
beta_0_vals = np.linspace(0, 10, 50)
beta_1_vals = np.linspace(0, 4, 50)
B0, B1 = np.meshgrid(beta_0_vals, beta_1_vals)
MSE = np.zeros_like(B0)

# --- Compute MSE for each combination ---
for i in range(B0.shape[0]):
    for j in range(B0.shape[1]):
        MSE[i,j] = mse(B0[i,j], B1[i,j], X, Y)

# --- Plot 3D surface ---
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(B0, B1, MSE, cmap='viridis', alpha=0.8)
ax.set_xlabel("beta_0")
ax.set_ylabel("beta_1")
ax.set_zlabel("MSE")
ax.set_title("3D Error Surface (MSE) for Linear Regression")

# Optional: mark minimum MSE from analytical solution
b0_analytical, b1_analytical = calculate_coefficients(X, Y)
min_mse = mse(b0_analytical, b1_analytical, X, Y)
ax.scatter(b0_analytical, b1_analytical, min_mse, color='red', s=50, label='Analytical Min')
ax.legend()

plt.show()
```
