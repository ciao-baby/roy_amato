```python
# simulation.py
import asyncio
import json
import random
import csv
import time
from datetime import datetime
from typing import Dict, Any, List

# --- Helpers / Simple Agent Simulator ---
class PersonaStore:
    def __init__(self, path="personas.json"):
        with open(path, "r", encoding="utf-8") as f:
            self.personas = json.load(f)

    def sample(self, n=1):
        return random.sample(self.personas, n)

class Agent:
    def __init__(self, persona: Dict[str, Any], run_id: int, scenario: str):
        self.persona = persona
        self.run_id = run_id
        self.scenario = scenario
        self.start_time = None
        self.events = []

    async def perform_task(self, feature_artifact):
        # Simulate task flow with probabilistic branching based on persona
        self.start_time = time.time()
        p = self.persona["probability_model"]
        # Stage 1: onboarding
        await asyncio.sleep(random.uniform(0.05, 0.25))  # simulated latency
        self.events.append(("onboarding_viewed", time.time()))
        # If low patience, might abandon early
        if random.random() > self.persona["behavioral_traits"]["patience"] / 10.0:
            if random.random() > p.get("tries_second_time", 0.5):
                self.events.append(("abandoned_early", time.time()))
                return {"status": "abandoned", "reason": "low_patience", "events": self.events}
        # Stage 2: task execution (simulate steps)
        steps = feature_artifact.get("steps", 3)
        for step in range(1, steps + 1):
            # time per step influenced by tech proficiency and device
            base = random.uniform(0.2, 0.7)
            speed_factor = (10 - self.persona["tech_proficiency"]) / 10.0
            await asyncio.sleep(base * (1.0 + speed_factor))
            self.events.append((f"step_{step}_completed", time.time()))
            # chance to be blocked/confused
            if random.random() < (1 - self.persona["behavioral_traits"]["curiosity"] / 10.0) * 0.1:
                self.events.append(("confused", time.time()))
        # Stage 3: finalization
        total_time = time.time() - self.start_time
        result = {
            "status": "completed",
            "total_time": round(total_time, 2),
            "events": self.events,
            "persona_id": self.persona["id"],
            "persona_name": self.persona["name"],
            "run_id": self.run_id,
            "scenario": self.scenario,
            "timestamp": datetime.utcnow().isoformat() + "Z"
        }
        return result

class Simulation:
    def __init__(self, persona_store: PersonaStore, concurrency: int = 10):
        self.store = persona_store
        self.concurrency = concurrency
        self.results = []

    async def run_one(self, persona, run_id, feature_artifact, scenario):
        agent = Agent(persona, run_id, scenario)
        try:
            res = await agent.perform_task(feature_artifact)
        except Exception as exc:
            res = {"status": "error", "error": str(exc), "persona_id": persona["id"], "run_id": run_id}
        return res

    async def run_parallel(self, personas: List[Dict], feature_artifact: Dict, scenario: str, runs_per_persona=5):
        sem = asyncio.Semaphore(self.concurrency)
        async def _worker(persona, run_id):
            async with sem:
                return await self.run_one(persona, run_id, feature_artifact, scenario)

        tasks = []
        run_counter = 0
        for persona in personas:
            for i in range(runs_per_persona):
                run_counter += 1
                tasks.append(asyncio.create_task(_worker(persona, run_counter)))
        for t in asyncio.as_completed(tasks):
            res = await t
            self.results.append(res)
        return self.results

    def dump_results_csv(self, path="simulation_results.csv"):
        keys = ["run_id", "persona_id", "persona_name", "scenario", "status", "total_time", "timestamp"]
        with open(path, "w", newline="", encoding="utf-8") as f:
            w = csv.DictWriter(f, fieldnames=keys)
            w.writeheader()
            for r in self.results:
                w.writerow({
                    "run_id": r.get("run_id"),
                    "persona_id": r.get("persona_id"),
                    "persona_name": r.get("persona_name"),
                    "scenario": r.get("scenario"),
                    "status": r.get("status"),
                    "total_time": r.get("total_time", ""),
                    "timestamp": r.get("timestamp", "")
                })

# --- Example usage: run a simple sim ---
if __name__ == "__main__":
    async def main():
        store = PersonaStore("personas.json")
        # sample personas to run (all)
        personas = store.personas
        sim = Simulation(store, concurrency=20)
        feature_artifact = {"name": "onboarding_flow_v1", "steps": 4}
        results = await sim.run_parallel(personas, feature_artifact, scenario="task_completion", runs_per_persona=10)
        sim.dump_results_csv("simulation_results.csv")
        # also write results json
        with open("simulation_results.json", "w", encoding="utf-8") as f:
            json.dump(results, f, indent=2)
        print("Done. Wrote simulation_results.csv and simulation_results.json")

    asyncio.run(main())
```

```python
# api/server.py
from fastapi import FastAPI, BackgroundTasks
from pydantic import BaseModel
import asyncio
import json
from simulation import Simulation, PersonaStore

app = FastAPI(title="TinyTroup Simulation API")
store = PersonaStore("personas.json")
_sim = Simulation(store, concurrency=20)

class RunRequest(BaseModel):
    scenario: str = "task_completion"
    runs_per_persona: int = 5
    concurrency: int = 20
    feature_artifact: dict = {"name": "onboarding_flow_v1", "steps": 3}

@app.post("/run")
async def start_run(req: RunRequest, background_tasks: BackgroundTasks):
    _sim.concurrency = req.concurrency
    # Run in background to avoid blocking the request; returns run_id immediately
    async def bg():
        await _sim.run_parallel(store.personas, req.feature_artifact, req.scenario, runs_per_persona=req.runs_per_persona)
        _sim.dump_results_csv("simulation_results.csv")
        with open("simulation_results.json", "w", encoding="utf-8") as f:
            json.dump(_sim.results, f, indent=2)
    background_tasks.add_task(asyncio.create_task, bg())
    return {"status": "accepted", "message": "Simulation started in background"}

@app.get("/results")
def get_results():
    try:
        with open("simulation_results.json", "r", encoding="utf-8") as f:
            data = json.load(f)
    except FileNotFoundError:
        data = []
    return {"count": len(data), "results": data}
```
