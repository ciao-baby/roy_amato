```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(
    "/content/2020-2025.csv",
    usecols=["Country", "2020"]
)

# Get top 10 by highest 2020 values
top10 = df.sort_values(by="2020", ascending=False).head(10)

# Plot bar chart
top10.plot(kind="bar", x="Country", y="2020", figsize=(10,6), legend=False)
plt.ylabel("2020 Value")
plt.title("Top 10 Countries by 2020 Value")
plt.xticks(rotation=45)
plt.show()
```
